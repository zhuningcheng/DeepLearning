```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import LabelBinarizer
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_extraction import DictVectorizer
```

```python
# 数据集中的education一列可以被视为是分类型数据集，教育划分为三个等级，内在是有顺序关系的特征，是ordinal。而marital可以视为norminal，已婚情况就分成两类：
# 未婚和已婚
dataframe = pd.read_csv("数据集/bank/bank.csv", header = 0, sep=';')
```

```python
test2 = np.array(dataframe["education"])
test2 = test2.reshape(-1,1)

# ordinal类型的编码方式可以使用OrdinalEncoder或者使用标签编码(label encoding)

# 使用OrdinalEncoder编码
# 每一个类别的编码从 0 ~ n_classes-1，编码器如果不进行设置categories时，按照默认顺序进行排序，可能在某些情况下会出错。

# 对于有unknown这种未知类型的,要结合handle_unknown和unknown_value一起使用。handle_unknown默认值是'error'直接报错，换成'use_encoded_value'并结合unknown_value
# 就不会报错了，unknown_value的值必须是整形或np.nan。encoded_missing_value用于处理缺失值问题

# 创建OrdinalEncoder编码器
encoder = OrdinalEncoder(categories = [["primary","secondary","tertiary"]],unknown_value=np.nan,handle_unknown="use_encoded_value", encoded_missing_value=np.nan)

# 使用对象方法
# 进行编码,创建编码后的特征矩阵
encoded_features = encoder.fit_transform(test2[500:510])
encoded_features

# 输出特征的所有取值（类别）
encoder.categories_

# 逆向编码，输出原本的表示
encoder.inverse_transform(encoded_features)
```

```python
test2 = np.array(dataframe["education"])
# 使用标签编码(label encoding)
# 和OrdinalEncoder编码一样，每一个类别的编码从 0 ~ n_classes-1,这里LabelEncoder()期望的是一个 “一维” 的矩阵，但是也是支持二维的矩阵，但是二维矩阵
# 的形状必须是n_features * 1的列向量的形状，不可以是 1 * n_features的行向量的形状 

# 创建对象
encoder = LabelEncoder()

# 使用对象方法,创建编码后的特征矩阵
encoded_features = encoder.fit_transform(test2)

encoded_features[0:10]
# 输出特征的所有取值（类别）
encoder.classes_

# 逆向编码
encoder.inverse_transform(encoded_features)[0:10]
```

```python
# 有些时候，样本特征使用字典来表示，一个字典表示一个样本，n个字典表示有n个样本。字典中的键值对表示样本的特征
# DictVectorizer()对象传入的数据是一个字典序列

# 随意构建一个字典序列
# 可以看到，样本中一共有三种不同的颜色，所以构建的特征向量长度就是3，特征向量长度=不同的关键词(key)的个数
data_list=[{'red':1,'blue':4},
           {'yellow':3,'blue':1},
          {'red':2,'green':1},
          {'blue':2,'yellow':1}]

# 创建对象
encoder = DictVectorizer(sparse = False)

# 进行转换
features  = encoder.fit_transform(data_list)
features

# 获取所有的key的值
encoder.get_feature_names_out(features)
#help(DictVectorizer)

# 逆向编码
encoder.inverse_transform(features)
```

```python
# 不平衡分类的处理，数据分布极其的不均衡，某种类别的数量远超于其他的数量
# 样本不平衡问题,处理方式有两种，（1）采用上采样方式：指的是对于样本中数量少的一类，我们通过某种方式增加该类的数量，使得两者的数量达到平衡
# （2）下采样：指的是对于样本中数量多的一类，我们通过某种方式减少该类的数量，使得两者的数量达到平衡

# 上采样方法中有（1）随机上采样（2）SMOTE（3）自适应上采样等，这里采用了SMOTE
from imblearn.over_sampling import SMOTE
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt

# X是特征矩阵，y是特征向量
X, y = make_classification(n_classes=2, class_sep=2,
                            weights=[0.1, 0.9], n_informative=3, n_redundant=1,
                            flip_y=0, n_features=20, n_clusters_per_class=1,
                            n_samples=1000, random_state=42)
# 创建SMOTE对象,有一个重要参数：k_neighbors表示使用几个邻居进行插值，默认是5个
smote = SMOTE(random_state=42)

# 使用对象方法

# 进行插值,fit_resample(X,y),X是原始样本的特征矩阵，y是原始的特征标签,返回有两个结果，X_resample:上采样后的特征矩阵，Y_resample:上采样后的特征向量
X_resampled , y_resampled = smote.fit_resample(X,y)

# 画出采样前和采样后两个类别的数量对比
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.scatter(X[y == 0][:, 0], X[y == 0][:, 1], label='Class 0', alpha=0.5)
plt.scatter(X[y == 1][:, 0], X[y == 1][:, 1], label='Class 1', alpha=0.5)
plt.title('Original dataset')
plt.legend()

plt.subplot(1, 2, 2)
plt.scatter(X_resampled[y_resampled == 0][:, 0], X_resampled[y_resampled == 0][:, 1], label='Class 0', alpha=0.5)
plt.scatter(X_resampled[y_resampled == 1][:, 0], X_resampled[y_resampled == 1][:, 1], label='Class 1', alpha=0.5)
plt.title('Resampled dataset with SMOTE')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# 这里再使用下采样方法，这里就使用的随即下采样的方式
from imblearn.under_sampling import RandomUnderSampler

# 创建仿真数据
X, y = make_classification(n_classes=2, class_sep=2,
                            weights=[0.1, 0.9], n_informative=3, n_redundant=1,
                            flip_y=0, n_features=20, n_clusters_per_class=1,
                            n_samples=1000, random_state=42)
# 创建对象
RUS = RandomUnderSampler()

# 使用对象方法
X_resampled , y_resampled = RUS.fit_resample(X,y)

# 画出采样前和采样后两个类别的数量对比
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.scatter(X[y == 0][:, 0], X[y == 0][:, 1], label='Class 0', alpha=0.5)
plt.scatter(X[y == 1][:, 0], X[y == 1][:, 1], label='Class 1', alpha=0.5)
plt.title('Original dataset')
plt.legend()

plt.subplot(1, 2, 2)
plt.scatter(X_resampled[y_resampled == 0][:, 0], X_resampled[y_resampled == 0][:, 1], label='Class 0', alpha=0.5)
plt.scatter(X_resampled[y_resampled == 1][:, 0], X_resampled[y_resampled == 1][:, 1], label='Class 1', alpha=0.5)
plt.title('Resampled dataset with SMOTE')
plt.legend()

plt.tight_layout()
plt.show()
```
