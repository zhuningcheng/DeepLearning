```python
# 散点图、多个子图的画法——使用的是subplots()

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("D:/PycharmProjects/pyjupyter/数据集/wine_quality/winequality-white.csv", sep=';')
columns = data.columns

fig, axes = plt.subplots(5, 2, figsize=(20, 20), dpi=100)
axes = axes.flatten()  # 将5x2的axes数组展平为1维，必须展开成一维才可以直接使用axes[i]
i = 0
for x in range(5):
    for y in range(x + 1, 5):
        if columns[x] == columns[y]:
            continue
        if i >= len(axes):  # 防止超出子图数量
            break
            
        X = data[columns[x]][:1000] # 取前三千行查看
        Y = data[columns[y]][:1000]
        axes[i].scatter(X, Y, alpha=0.4, label='white wine')
        axes[i].set_xlabel(columns[x])
        axes[i].set_ylabel(columns[y])
        axes[i].legend(loc="best")
        i += 1

plt.show()
```

```python
# 折线图、直方图、柱状图、饼图、多个子图的画法——使用的是add_subplot()
data = pd.read_csv("D:/PycharmProjects/pyjupyter/数据集/wine_quality/winequality-white.csv", sep=';')
columns = data.columns

fig = plt.figure()  # 先生成一个主画布，再通过Figure对象调用add_subplot()方法
ax1 = fig.add_subplot(2, 2, 1)  # nrow:画布的行数,ncol:表示画布的列数,index:表示子画布的编号，从1开始计算
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)
ax4 = fig.add_subplot(2, 2, 4)

ax1.set_xticks(data[columns[0]][:5])  # 设置x轴刻度，可以用xticks或set_xticks()
x1 = data[columns[0]][:5]
y1 = data[columns[1]][:5]
ax1.set_label(columns[0])
ax1.set_label(columns[1])
ax1.scatter(x1, y1, label='white wine')
ax1.legend(loc="best")
ax1.annotate('3 and 4',xy=(7.2, 0.23), xytext=(7.2, 0.24), arrowprops=dict(arrowstyle='-|>',color='red'))
plt.show()
```

```python
# 折线图、直方图、柱状图、饼图画法
data = pd.read_csv("D:/PycharmProjects/pyjupyter/数据集/wine_quality/winequality-white.csv", sep=';')

# 折线图plot()
plt.figure()
X = data['pH'][:20]
Y = data['quality'][:20]
plt.xlabel('pH')
plt.ylabel('quality')
plt.sactter(X, Y ,label='white wine')
plt.legend(loc="best")
plt.show()
```

```python
# 柱状图
data = pd.read_csv("D:/PycharmProjects/pyjupyter/数据集/wine_quality/winequality-white.csv", sep=';')

# 折线图bar()
plt.figure()
X = data['pH'][:20]
Y = data['quality'][:20]
plt.xlabel('pH')
plt.ylabel('quality')
plt.bar(X, Y ,label='white wine')
plt.legend(loc="best")
plt.show()
```

```python
# 直方图
data = pd.read_csv("D:/PycharmProjects/pyjupyter/数据集/wine_quality/winequality-white.csv", sep=';')

# 折线图hist()
plt.figure()
X = data['pH'][:20]
Y = data['quality'][:20]
plt.xlabel('pH')
plt.ylabel('quality')
plt.hist(X,bins=None,label='white wine')
plt.legend(loc="best")
plt.show()
```

```python
# 饼图
data = pd.read_csv("D:/PycharmProjects/pyjupyter/数据集/wine_quality/winequality-white.csv", sep=';')

# 饼图pie()
plt.figure()
X = data['pH'][:20]
Y = data['quality'][:20]
plt.pie(X,autopct='')
plt.show()
```
